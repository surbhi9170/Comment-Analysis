```python
import os
import re
import tweepy 
from textblob import TextBlob
```

```python
import pandas as pd
import numpy as np
```

```python
tweet_df=pd.read_excel('Assignment for ML engineer.xlsx', sheet_name=3)
tweet_df.head(20)
```

```python
a=list(tweet_df['Comment'])
```

```python
def clean_tweet(tweet):
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split()) 
```

```python
tweet_details = [TextBlob(clean_tweet(tweet)).sentiment.polarity for tweet in a]
```

```python
tweet_df['cleaned comments']=[clean_tweet(tweet) for tweet in a]
tweet_df['sentiment']=[float(i) for i in tweet_details]
```

```python
tweet_df.head(20)
```

```python
#tweet_df['sentiment'] = tweet_df['sentiment'].apply(lambda x: 1 if x >0 else 0)
tweet_df.loc[tweet_df['sentiment'] == 0 , 'sentiment d'] = 'neutral'
tweet_df.loc[tweet_df['sentiment'] > 0, 'sentiment d'] = 'positive'
tweet_df.loc[tweet_df['sentiment'] < 0 , 'sentiment d'] = 'negative'
```

```python
tweet_df['sentiment d'].value_counts()
```

```python
tweet_df.head(20)
```

```python
s=sum(tweet_details)
if s<0:
    print("comment analysis says that the post has negative sentiment")
elif s==0:
    print("comment analysis says that the post has neutral sentiment")
elif s>0:
    print("comment analysis says that the post has positive sentiment")
```
