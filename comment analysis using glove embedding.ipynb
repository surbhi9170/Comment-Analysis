```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten,Embedding, Activation
from tensorflow.keras.layers import Conv1D , MaxPooling1D, GlobalMaxPooling1D, Dropout
from tensorflow.keras import regularizers
import numpy as np
from numpy import array
from sklearn.model_selection import train_test_split
```

```python
text = tweet_df['tweets'].tolist()
text
```

```python
y=tweet_df['sentiment']
```

```python
token = Tokenizer(filters='!"#$%&()*+,-./:;<=>?@[\]^_`{"}~\t\n🙏🚨◘📣⚠️🌏🤔✅☣️️⭐',lower=True, split=" ")
token.fit_on_texts(text)
```

```python
token
```

```python
vocab_size = len(token.word_index)+1
vocab_size
```

```python
token.index_word
```

```python
x=['my health is fine']
```

```python
token.texts_to_sequences(x)
```

```python
encoded_text = token.texts_to_sequences(text)
```

```python
(encoded_text)
```

```python
max_length = 120
X = pad_sequences(encoded_text, maxlen=max_length, padding='post')
```

```python
print(X)
```

```python
X.shape
```

```python
X = np.array(X)
y = np.array(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42,test_size=0.2, stratify=y)
```

```python
vec_size = 300

model = Sequential()
model.add(Embedding(vocab_size, vec_size, input_length = max_length))

model.add(Conv1D(64, 8, activation='relu'))
model.add(MaxPooling1D(2))
model.add(Dropout(0.2))

model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))


model.add(Dense(16, activation='relu'))

model.add(GlobalMaxPooling1D())

model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss= 'binary_crossentropy', metrics=['accuracy'])
```

```python
%%time
model.fit(X_train, y_train, epochs = 5, validation_data= (X_test, y_test))
```

```python
def get_encoded(x):
    x= token.texts_to_sequences(x)
    x=pad_sequences(x, maxlen=max_length, padding = 'post')
    return x
```

```python
from pathlib import Path
emb_dict = {}
glove = open(r'D:\7th SEMESTER\NTCC\glove.twitter.27B.50d.txt', encoding="utf8")
for line in glove:
    values = line.split()
    word = values[0]
    vector = np.asarray(values[1:], dtype='float32')
    emb_dict[word] = vector
glove.close()
```

```python
print('Loaded %s word vectors.' % len(emb_dict))
```

```python
emb_matrix = np.zeros((vocab_size, 50))

for w, i in token.word_index.items():
    # The word_index contains a token for all words of the training data so we need to limit that
    if i < vocab_size:
        vect = emb_dict.get(w)
        # Check if the word from the training data occurs in the GloVe word embeddings
        # Otherwise the vector is kept with only zeros
        if vect is not None:
            emb_matrix[i] = vect
    else:
        break
```

```python
glove_model = Sequential()
glove_model.add(Embedding(vocab_size,50, input_length=max_length))
glove_model.add(Flatten())
glove_model.add(Dense(1, activation='sigmoid'))
glove_model.summary()
```

```python
glove_model.compile(optimizer='adam', loss= 'binary_crossentropy', metrics=['accuracy'])
```

```python
%%time
glove_model.fit(X_train, y_train, epochs = 5, validation_data= (X_test, y_test))
```
